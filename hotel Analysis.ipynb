```python
#Canceled bookings
canceled_perc = df['is_canceled'].value_counts(normalize = True)
print(cancelled_perc)
plt.figure(figsize =(5,4) )
plt.title('Reservation status count')
plt.bar(['Not canceled','Canceled'],df['is_canceled'].value_counts(),edgecolor = 'k',width = 0.7)
plt.show()
```

```python
plt.figure(figsize = (8,4))
ax1= sns.countplot(x = 'hotel',hue = 'is_canceled',data = df,palette  = 'Blues')
legend_labels,_ = ax1.get_legend_handles_labels()
ax1.legend(title="Reservation Status", labels=["Not Canceled", "Canceled"], bbox_to_anchor=(1, 1), loc='upper left')
plt.title('Reservation status in different hotels',size = 20)
plt.xlabel('hotel')
plt.ylabel('number of reservations')
plt.legend(['not canceled','canceled'])
plt.show()
```

```python
resort_hotel = df[df['hotel'] == 'Resort hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
city_hotel = df[df['hotel'] =='city hotel']
city_hotel['is_canceled'].value_counts(normalize = True)
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')['adr'].mean()
city_hotel   = city_hotel.groupby('reservation_status_date')['adr'].mean()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(20,8))
plt.title('Average Daily Rate in City and Resort Hotel', fontsize=30)
plt.plot(resort_hotel.index, resort_hotel.values, label='Resort Hotel')
plt.plot(city_hotel.index, city_hotel.values, label='City Hotel')
plt.legend(fontsize=20)
plt.show()

```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

 
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'], errors='coerce')

 
df['month'] = df['reservation_status_date'].dt.month

 
plt.figure(figsize=(16,8))
ax1 = sns.countplot(x='month', hue='is_canceled', data=df, palette='bright')
 
ax1.legend(title="Reservation Status",
           labels=["Not Canceled", "Canceled"],
           bbox_to_anchor=(1,1))

plt.title('Reservation Status per Month', fontsize=20)
plt.xlabel('Month')
plt.ylabel('Number of Reservations')

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(15,8))
plt.title('ADR per Month', fontsize=30)

sns.barplot(x='month', y='adr', data=df, palette='viridis')

plt.xlabel('Month', fontsize=20)
plt.ylabel('Average Daily Rate (ADR)', fontsize=20)
plt.show()
```

```python
plt.figure(figsize = (15,8))
plt.title('ADR per Month', fontsize=30)
sns.barplot(x='month', y='adr', data=df[df['is_canceled']==1].groupby('month')[['adr']].sum().reset_index())
plt.xlabel('Month', fontsize=20)
plt.ylabel('Average Daily Rate (ADR)', fontsize=20)
plt.show()
```

```python
# Create a 'year' column from the reservation date
df['year'] = pd.to_datetime(df['reservation_status_date']).dt.year

# Plot ADR per year for canceled bookings
plt.figure(figsize=(15,8))
plt.title('ADR per Year', fontsize=30)
sns.barplot(
    x='year', 
    y='adr', 
    data=df[df['is_canceled']==1].groupby('year')[['adr']].sum().reset_index()
)
plt.xlabel('Year', fontsize=20)
plt.ylabel('Average Daily Rate (ADR)', fontsize=20)
plt.show()
```

```python
# Filter canceled reservations
cancelled_date = df[df['is_canceled'] == 1]

# Get top 10 countries by canceled reservations
top_10_country = cancelled_date['country'].value_counts()[:10]

# Plot pie chart
plt.figure(figsize=(8,8))
plt.title('Top 10 Countries with Reservations Canceled', fontsize=16)
plt.pie(top_10_country, autopct='%.2f%%', labels=top_10_country.index)
plt.show()
```

```python

# Calculate average ADR per date for canceled bookings
cancelled_df_adr = cancelled_date.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace=True)
cancelled_df_adr.sort_values('reservation_status_date', inplace=True)

# Calculate average ADR per date for not canceled bookings
not_cancelled_df_adr = df[df['is_canceled']==0].groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace=True)
not_cancelled_df_adr.sort_values('reservation_status_date', inplace=True)

# Plot ADR over time
plt.figure(figsize=(20,6))
plt.title('Average Daily Rate Over Time', fontsize=18)
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label='Not Cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label='Cancelled')
plt.xlabel('Date', fontsize=14)
plt.ylabel('Average Daily Rate (ADR)', fontsize=14)
plt.legend()
plt.show()
```
